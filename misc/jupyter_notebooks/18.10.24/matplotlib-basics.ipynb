```python
%autosave 20
# %matplotlib notebook
%matplotlib inline
import matplotlib
# matplotlib.use('Agg') - не открывать окон
import matplotlib.pyplot as plt
# from matplotlib import pyplot as plt

import numpy as np
```

```python
x = np.r_[0:2*np.pi:11j]
y = np.sin(x)
plt.figure(figsize=(10, 8))
plt.title('sin')
plt.xlabel(r'$\xi$')
plt.ylabel('$y$')
plt.xticks((0, 1, 5, 6), ('zero', 'I', 'V', 'VI'))
plt.plot(x, y, '*', color=[0.5, 0.3, 0.6], label='data')
plt.plot(x, y**2, '--', lw=5, color='blue', label=r'sin$^2$')
plt.text(1, -0.5, 'text')
# plt.hlines()
plt.legend(loc='best')
plt.grid()
plt.savefig('sin.eps')
plt.savefig('sin.pdf')
plt.savefig('sin.png')
```
